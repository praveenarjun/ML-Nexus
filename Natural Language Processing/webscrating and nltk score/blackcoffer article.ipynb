# **writing all process in single call back fuction**

```python
#saving blank url saparate set
Blank_url={}
def web_scraping(self):
  data = pd.read_excel('/content/Input.xlsx')
  # Create a copy from the original DataFrame
  df = data.copy()

  # Create an 'article_words' column
  df['article_words'] = ' '

  for i, url in enumerate(df['URL']):
    response_code = requests.get(url)
    soup = bs(response_code.text, 'html.parser')
    article_t = soup.find('title').text

    all_text_element = soup.find("div", class_="td-pb-span8 td-main-content")

    #using if-else to find which all url has [request=200]
    if all_text_element is not None:
      all_text = all_text_element.get_text(strip=True, separator='\n')
      firstdata =all_text.splitlines()
    else:
      print (f"No text found in the HTML for URL: {url}")
      Blank_url[f"blankassign00{i+1}"] = url
      firstdata = []

    filename = urllib.parse.quote_plus(url)
    file_path = '/content/new url.txt'
    space = ' '

    #ceating new files
    with open(f"{file_path}\ {filename}.txt", 'w') as file1:
        file1.writelines (article_t)
        file1.writelines (space)
        file1.writelines (firstdata)

    # new 'article_words' column
    df.at[i, 'article_words'] = f"{article_t}-{firstdata}"
  df.to_csv('/content/final.csv', index=False)
  return df
```

```python
#call the fuction web_scraping
new_data = web_scraping(Data)
```

```python
import re
#fuction to tokenized and lemminatized text
def text_saparater(x):
  string_format = str(x).lower()
  lower_words=re.sub('[^a-zA-Z]+',' ',string_format).strip()
  #lower_words = lower_words.split()
  token = word_tokenize(lower_words)
  token_word = [t for t in token if t not in (stopword_Auditor, stopword_Currencies, stopword_DatesandNumbers, stopword_Generic, stopword_Genericlong, stopword_Geographic, stopword_Names)]
  lem = WordNetLemmatizer()
  lemantizzed = [lem.lemmatize(w) for w in token_word]
  return lemantizzed
```

```python
def Personal_pronoun_count(words_list):
    list_of_words = ['I', 'we', 'my' ,'ours','us' ]
    list_words_counts = 0
    for words in words_list:
        if words in list_of_words:
            list_words_counts += 1
    return list_words_counts

Personal_pronoun_count(words)
```
